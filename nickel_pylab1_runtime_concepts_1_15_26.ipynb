```python
a = 10
b = a
a = 20
print(b)
```

Was the object mutable or immutable?  
- a and b are both immutable.  

Did Python mutate the object or rebind the name?  
- Python rebinded a.



```python
def f(x):
  x = x + 1
  print(x)

n = 5
f(n)
print(n)
```

Was the object mutable or immutable?  
- Both x and n are both immutable because they are integers.   

Did Python mutate the object or rebind the name?  
- n was not rebinded or mutated. x was rebinded locally in f(x).


```python
x = 5
print(id(x))
x += 1
print(id(x))
```

Was the object mutable or immutable?  
-   x is immutable because it is an integer.

Did Python mutate the object or rebind the name?  
- Python rebinded x. This is because the variable x keeps the same name but refers to a different object in memory.



```python
lst = [1,2]
print(id(lst))
lst.append(3)
print(id(lst))
```

Was the object mutable or immutable?  
- lst is a mutable object because it is a list.  

Did Python mutate the object or rebind the name?  
- Python mutated the object using lst.append(3).



```python
def change(x):
  x = x + 10

a = 5
change(a)
print(a)
```

Was the object mutable or immutable?  
- Both a and x are immutable because they are integers.  

Did Python mutate the object or rebind the name?  
- Python did not rebind a. Python rebinded x locally inside the function.


```python
def change(lst):
  lst.append(99)

data = [1,2]
change(data)
print(data)
```

Was the object mutable or immutable?  
- data and lst are mutable objects.   

Did Python mutate the object or rebind the name?  
- Python mutated both objects.



```python
def reset_wrong(lst):
  lst = []

def reset_right(lst):
  lst.clear()

data = [1,2,3]
reset_wrong(data)
print(data)

reset_right(data)
print(data)
```

Was the object mutable or immutable?  
- data and lst are mutable objects.  

Did Python mutate the object or rebind the name?  
- In reset_wrong(), the lst was rebinded locally. In resest_right(), lst was mutated. Using reset_right(), data was mutated.



```python
def demo(x):
  print(id(x))
  x.append(5)
  print(id(x))

lst = [1,2]
print(id(lst))
demo(lst)
```

Was the object mutable or immutable?  
- lst and x are mutable objects.  

Did Python mutate the object or rebind the name?  
- Python mutated the list.
